```python
import numpy as np
```

```python
#designing a class perceptron
class Percep:

    def __init__(self, lr=0.1, ep=20):
        # lr = learning rate
        # ep = number of training epochs
        self.lr = lr
        self.ep = ep

        # init weights and bias
        self.w = np.zeros(2)
        self.b = 0

    def step(self, z):
        # step activation
        return 1 if z >= 0 else 0

    def pred(self, x):
        # calc score
        z = np.dot(self.w, x) + self.b
        return self.step(z)

    def train(self, X, y):

        for _ in range(self.ep):
            for i in range(len(X)):

                x = X[i]
                t = y[i]

                # prediction
                y_hat = self.pred(x)

                # error
                err = t - y_hat

                # update rule
                self.w += self.lr * err * x
                self.b += self.lr * err
```

```python
def get_data(gate):

    X = np.array([
        [0,0],
        [0,1],
        [1,0],
        [1,1]
    ])

    if gate == "AND":
        y = np.array([0,0,0,1])

    elif gate == "OR":
        y = np.array([0,1,1,1])

    elif gate == "NAND":
        y = np.array([1,1,1,0])

    elif gate == "NOR":
        y = np.array([1,0,0,0])

    elif gate == "XOR":
        y = np.array([0,1,1,0])

    return X, y
```

```python
def run_gate(g):

    X, y = get_data(g)

    p = Percep(lr=0.1, ep=20)
    p.train(X, y)

    print("\nGate:", g)
    print("Weights:", p.w)
    print("Bias:", p.b)

    print("Predictions")

    for x in X:
        print(x, "->", p.pred(x))
```

```python
gates = ["AND","OR","NAND","NOR","XOR"]

for g in gates:
    run_gate(g)
```

# 1. Effect of Learning Rate

The learning rate determines how much the model updates its weights after each error during training.

a. Small learning rate: The model learns slowly but the training process is more stable and gradually moves toward convergence.

b. Large learning rate: The model learns faster, but large updates may cause the training process to overshoot the optimal solution, leading to oscillations or failure to converge.

In this experiment, using a moderate learning rate such as 0.1 provides a good balance, allowing the perceptron to train steadily for linearly separable logic gates.

# 2. Why the Same Code Learns Different Gates

The structure of the perceptron model remains unchanged. The only difference is the training dataset used for each logic gate.

Each dataset represents a different input output relationship. During training, the learning algorithm adjusts the weights and bias based on this data.

As a result, even though the code is the same, the model learns different decision boundaries for each gate because the data guides how the parameters are updated.
